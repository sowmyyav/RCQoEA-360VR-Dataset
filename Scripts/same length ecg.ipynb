```python
import pandas as pd 
import matplotlib.pyplot as plt
import joblib
import csv

# Load the CSV file into a Pandas DataFrame
ecg_data_all = pd.read_csv('/home/sowmya/QOE_CWI/Selected_PID_data/ECG_merged_data_1.csv', header=None)
# Assuming ecg_data_all is a list/array containing ECG data for multiple participants

# Display the first few rows of the DataFrame to verify the data has been loaded
#print(ecg_data_all.head())

#num_participants = len(ecg_data_all)

desired_length = 3768
# Function to trim NaN values beyond the desired length in each row
def trim_nan_values(row):
    if len(row) > desired_length:
        return row[:desired_length]
    else:
        return row

# Apply the function to each row in the DataFrame
df_trimmed = ecg_data_all.apply(trim_nan_values, axis=1)

# Create the filename
filename = f"/home/sowmya/QOE_CWI/Selected_PID_data/ECG_equal_length_data.csv"

# Write the DataFrame to a CSV file
df_trimmed.to_csv(filename, index=False)        
# Write the filtered data to a CSV file

          
```

```python
plt.figure(figsize=(12, 8))

for i in range(8):
    plt.subplot(8, 1, i+1)
    plt.plot(ecg_data_all[i])
    plt.title(f'Participant {i+1} - ECG Signal')
    plt.xlabel('Time')
    plt.ylabel('ECG Signal Values')
    plt.grid(True)


plt.show()
```
