```python
import pandas as pd
import os

# Path to the directory containing your ECG data CSV files
directory_path = '/home/sowmya/QOE_CWI/Selected_PID_data/GSR_data_all'

# Get a list of all files in the directory
#file_list = os.listdir(directory_path)

# Initialize an empty list to store ECG data
merged_gsr_data = []

# Get a list of CSV files in the folder
csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]

# Define a custom sorting function to sort the files appropriately
def sort_key(file_name):
    # Split the file name into parts
    parts = file_name.split('_')
    num1_part = parts[-2][1:]  # Extract the numerical part after the first '_'
    num2_part = parts[-1].split('.')[0]  # Extract the numerical part before '.csv'
    return (int(num1_part), int(num2_part))  # Convert both parts to integers for sorting

# Sort the list of file names using the custom sorting function
csv_files.sort(key=sort_key)

# Loop through each file in the directory
for file_name in csv_files:
    file_path = os.path.join(directory_path, file_name)
       
    print(file_path)
    # Read data from each CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)
    # Extract only the Shimmer_F5D4_GSR_Skin_Resistance_CAL data column (it's the tenth column)
    gsr_data = df.iloc[:, 16]  # Change the index if ECG data is in a different column
    # Append the ECG data to the merged list
    merged_gsr_data.append(gsr_data)

# Concatenate all ECG data into a single DataFrame
merged_gsr_df = pd.concat(merged_gsr_data, axis=1)
# Transpose the DataFrame
merged_gsr_df = merged_gsr_df.transpose()

# Export the merged data to a new CSV file
output_file_path = '/home/sowmya/QOE_CWI/Selected_PID_data/GSR_SC_merged_all_data.csv'
merged_gsr_df.to_csv(output_file_path, index=False, header=False)  # Exclude header for ECG data

print("Merged gsr data saved to", output_file_path)
```
