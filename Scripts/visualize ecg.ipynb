```python
import pandas as pd 
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
ecg_data_all = pd.read_csv('C:/Users/Sowmya/QOE_CWI/Selected_PID_data/ECG_merged_data_no_transpose.csv', header=None)
# Assuming ecg_data_all is a list/array containing ECG data for multiple participants

# Display the first few rows of the DataFrame to verify the data has been loaded
print(ecg_data_all.head())

num_participants = len(ecg_data_all)

plt.figure(figsize=(12, 8))

for i in range(num_participants):
    plt.plot(ecg_data_all[i], label=f'Participant {i+1}')  # Plot each participant's ECG data

plt.title('ECG Signals of Multiple Participants')
plt.xlabel('Time')
plt.ylabel('ECG Signal Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i in range(8):
    plt.subplot(8, 1, i+1)
    plt.plot(ecg_data_all[i])
    plt.title(f'Participant {i+1} - ECG Signal')
    plt.xlabel('Time')
    plt.ylabel('ECG Signal Values')
    plt.grid(True)


plt.show()
```
